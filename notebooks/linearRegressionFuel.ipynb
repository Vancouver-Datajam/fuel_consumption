```python
import pandas as pd
import os
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
```

```python
all_types = pd.read_csv('all_types.csv')
```

```python
all_types['fuel'].unique()
```

```python
gas_only = all_types.loc[all_types['fuel'] == 'gas' ]
```

```python
fit = stats.linregress(gas_only['co2_rating'], gas_only['engine_size'])
print("Fitted value of the co2_rating and engine_size:",
      "\n\tr2value -> ", str((fit.rvalue)**2),
     )
```

```python
sns.set(style="darkgrid")
sns.lmplot(x='engine_size', y='co2_rating', data=gas_only);
```

```python
gas_only_corr = gas_only.corr()
trimask = np.triu(np.ones_like(gas_only_corr, dtype=bool))

plt.figure(figsize=(10,6))
sns.heatmap(gas_only_corr, cmap='coolwarm', mask=trimask)
plt.show()
```
